```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from rake_nltk import Rake

# Download NLTK resources (if not already downloaded)
nltk.download('punkt')
nltk.download('wordnet')

# Initialize WordNet Lemmatizer
lemmatizer = WordNetLemmatizer()

# Input text
text = '''First off, you can buy two refurbished for a bit more than one new one. $120 vs $100. They sound killer. Get the sub. You’re gonna be happy. I suggest this model with the sub-they’re designed to match. The crossover between the Echo and the sub is seamless. Perfectly matched. No adjustment needed. The Alexa app is really well planned out. Plug the Echoes in, combine them to stereo, pick a room. Done. There’s been a ton of complaints about teo things. Random disconnects… if you’re like me, you might have 5 or 6 wifi devices in your room, with a wall or two between you and your router. It’s not a problem with the Echo-your router probably can’t handle the load. Check with your ISP for a newer model. With most of my gear on all over the place, even my ridiculously priced laptop runs at maybe half speed or less. Got a better router that could support 50 devices at once, problem solved. These run over wifi. I’ve only had a dropped connection problem once since i got them in February. Keep two of them a decent distance between, the seperation is amazing. ( side note-i don’t care what anybody tells you-don’t stick the sub behind something and right next to a wall. At least a foot away from a wall, and out in the open. The difference is really noticeable. I won’t bore you with details, look online. It absolutely needs that space)
'''
```

```python

# Tokenize the text
tokens = word_tokenize(text)

# Lemmatize the tokens
lemmatized_text = [lemmatizer.lemmatize(token) for token in tokens]

# Join the lemmatized tokens back into a string
lemmatized_text_str = ' '.join(lemmatized_text)

# Initialize RAKE
r = Rake()

# Extract keywords from the lemmatized text
r.extract_keywords_from_text(lemmatized_text_str)


# Get the ranked keywords
keywords = r.get_ranked_phrases()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample reviews
reviews = [
    "This product is amazing! I love the flavor and it's so easy to use.",
    "This product is not very good. The quality is low and it didn't meet my expectations.",
    "The product is okay, but I've seen better.",
]

# Preprocess reviews (optional)
# Remove stop words, punctuation, etc.

# Create TF-IDF vectorizer
vectorizer = TfidfVectorizer()
```

```python
# Fit and transform reviews
tfidf_matrix = vectorizer.fit_transform(reviews)

# Get feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Print TF-IDF scores for each word in each review
for i in range(len(reviews)):
    print(f"Review {i + 1}:")
    for word_index, word in enumerate(feature_names):
        print(f"{word}: {tfidf_matrix[i, word_index]}")
    print()
```

```python
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer

# Define stopwords (optional)
stopwords = nltk.corpus.stopwords.words('english')

# Define lemmatizer
lemmatizer = WordNetLemmatizer()

# Define function to preprocess text
def preprocess_text(text):
  # Lowercase text
  text = text.lower()
  # Tokenize text
  tokens = text.split()
  # Remove stopwords (optional)
  if stopwords:
    tokens = [token for token in tokens if token not in stopwords]
  # Lemmatize tokens
  tokens = [lemmatizer.lemmatize(token) for token in tokens]
  # Join tokens back into text
  processed_text = ' '.join(tokens)
  return processed_text

# Sample text
text = '''
First off, you can buy two refurbished for a bit more than one new one. $120 vs $100. They sound killer. Get the sub. You’re gonna be happy. I suggest this model with the sub-they’re designed to match. The crossover between the Echo and the sub is seamless. Perfectly matched. No adjustment needed. The Alexa app is really well planned out. Plug the Echoes in, combine them to stereo, pick a room. Done. There’s been a ton of complaints about teo things. Random disconnects… if you’re like me, you might have 5 or 6 wifi devices in your room, with a wall or two between you and your router. It’s not a problem with the Echo-your router probably can’t handle the load. Check with your ISP for a newer model. With most of my gear on all over the place, even my ridiculously priced laptop runs at maybe half speed or less. Got a better router that could support 50 devices at once, problem solved. These run over wifi. I’ve only had a dropped connection problem once since i got them in February. Keep two of them a decent distance between, the seperation is amazing. ( side note-i don’t care what anybody tells you-don’t stick the sub behind something and right next to a wall. At least a foot away from a wall, and out in the open. The difference is really noticeable. I won’t bore you with details, look online. It absolutely needs that space)
'''

# Preprocess text
processed_text = preprocess_text(text)

# Define TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform text
vectorized_text = vectorizer.fit_transform([processed_text])

# Print TF-IDF features and weights
features = vectorizer.get_feature_names_out()
weights = vectorized_text.toarray()[0]
for feature, weight in zip(features, weights):
  print(f"{feature}: {weight:.4f}")

```
